```python
# نصب کتابخانه‌های مورد نیاز
!pip install python-telegram-bot==13.15
!pip install python-docx
!pip install faiss-cpu
!pip install sentence-transformers
!pip install gradio_client

import os
from docx import Document
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from telegram import Update
from telegram.ext import Updater, CommandHandler, MessageHandler, Filters, CallbackContext
from gradio_client import Client

# تنظیمات اولیه
TELEGRAM_TOKEN = '7235681310:AAEIgJ8BhII-kvjNn7jZpMs2QqZ3Dkj5_Ac'    # جایگزین با توکن بات تلگرام شما

# تنظیمات مدل Gradio
client = Client("PartAI/Dorna-Llama3-8B-Instruct")

# تابع برای استخراج متن از فایل ورد
def extract_text(docx_path):
    doc = Document(docx_path)
    full_text = []
    for para in doc.paragraphs:
        if para.text.strip():
            full_text.append(para.text.strip())
    return '\n'.join(full_text)

# استخراج متن از همه فایل‌های ورد آپلود شده
documents = []
document_paths = [f for f in os.listdir('/content') if f.endswith('.docx')]

for doc_path in document_paths:
    full_path = os.path.join('/content', doc_path)
    text = extract_text(full_path)
    documents.append(text)

# ایجاد مدل بردارسازی چندزبانه
model_embed = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')

# ایجاد بردارهای معنایی از اسناد
embeddings = model_embed.encode(documents, convert_to_tensor=False)

# تبدیل بردارها به نوع float32 برای FAISS
embeddings = np.array(embeddings).astype('float32')

# ایجاد ایندکس FAISS
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

# ذخیره اسناد
doc_store = documents

# تابع برای بازیابی اسناد مرتبط
def retrieve_documents(query, top_k=1):
    query_embedding = model_embed.encode([query], convert_to_tensor=False)
    query_embedding = np.array(query_embedding).astype('float32')
    distances, indices = index.search(query_embedding, top_k)
    retrieved_docs = [doc_store[idx] for idx in indices[0]]
    return retrieved_docs

# تابع برای تولید پاسخ با استفاده از Gradio Client
def generate_response(retrieved_docs, question):
    context = " ".join(retrieved_docs)
    prompt = f"{context}\n\n{question}"

    try:
        result = client.predict(
            prompt,
            api_name="/chat"
        )
        answer = result.strip()
        if not answer:
            answer = "متاسفانه، نمی‌توانم به این سوال پاسخ دهم."
    except Exception as e:
        answer = "متاسفانه، در تولید پاسخ خطایی رخ داد."

    return answer

# هندلر برای شروع گفتگو
def start(update: Update, context: CallbackContext):
    update.message.reply_text("سلام! من یک بات هوش مصنوعی هستم. هر سوالی دارید بپرسید.")

# هندلر برای دریافت پیام‌های متنی
def handle_message(update: Update, context: CallbackContext):
    question = update.message.text
    if not question:
        update.message.reply_text("لطفاً یک سوال بپرسید.")
        return

    # بازیابی اسناد مرتبط
    retrieved_docs = retrieve_documents(question)

    # تولید پاسخ با استفاده از مدل
    answer = generate_response(retrieved_docs, question)

    # ارسال پاسخ به کاربر
    update.message.reply_text(answer)

# تابع اصلی برای اجرای بات
def main():
    # ایجاد Updater و Dispatcher با استفاده از روش Polling
    updater = Updater(TELEGRAM_TOKEN, use_context=True)
    dp = updater.dispatcher

    # افزودن هندلرها
    dp.add_handler(CommandHandler("start", start))
    dp.add_handler(MessageHandler(Filters.text & ~Filters.command, handle_message))

    # شروع بات با استفاده از Polling
    updater.start_polling()

    print("بات در حال اجرا است...")
    updater.idle()

if __name__ == '__main__':
    main()
```

```python
# نصب کتابخانه‌های مورد نیاز
!pip install python-telegram-bot==13.15
!pip install python-docx

import os
from telegram import Update, InlineKeyboardButton, InlineKeyboardMarkup
from telegram.ext import (
    Updater,
    CommandHandler,
    CallbackQueryHandler,
    CallbackContext,
    MessageHandler,
    Filters,
)

# تنظیمات اولیه
TELEGRAM_TOKEN = '7235681310:AAEIgJ8BhII-kvjNn7jZpMs2QqZ3Dkj5_Ac'    # جایگزین با توکن بات تلگرام شما

# تعریف دسته‌بندی‌ها
categories = {
    "c1": "مقررات کلی تحصیل",
    "c2": "نمرات، اعتراض و تجدیدنظر",
    "c3": "مرخصی و انصراف از تحصیل",
    "c4": "واحدهای درسی و حذف و اضافه",
    "c5": "امتحانات و پایان‌نامه",
    "c6": "انتقالی و تحصیل در دانشگاه‌های دیگر",
    "c7": "حذف و اضافه، واحدهای درسی و مشروط شدن",
    "c8": "فارغ‌التحصیلی و گواهینامه‌ها",
    "c9": "کارآموزی و دوره‌های عملی",
    "c10": "آزمون‌ها و تجدید نظر",
    "c11": "ثبت‌نام و مقررات اداری",
    "c12": "واحدهای جبرانی و دروس مهمان"
}

# تعریف سوالات و پاسخ‌ها
qa_pairs = {
    # دسته‌بندی ۱. مقررات کلی تحصیل
    "c1_q1": {
        "question": "هدف از تدوین این مقررات چیست؟",
        "answer": "هدف اصلی این مقررات، ایجاد یک چارچوب قانونی برای ارتقای کیفیت آموزش و ساماندهی وضعیت تحصیلی دانشجویان است."
    },
    "c1_q2": {
        "question": "حداقل نمره قبولی در هر درس چقدر است؟",
        "answer": "حداقل نمره قبولی برای هر درس ۱۰ است. اگر دانشجو نمره‌ای کمتر از این مقدار کسب کند، باید مجدداً درس را اخذ کند."
    },
    "c1_q3": {
        "question": "مدت زمان مجاز تحصیل در دوره کارشناسی چند سال است؟",
        "answer": "مدت زمان مجاز تحصیل در دوره کارشناسی ۴ سال است و در شرایط خاص، امکان افزایش این مدت تا ۵ سال وجود دارد."
    },
    "c1_q4": {
        "question": "شرایط فارغ‌التحصیلی در دوره کارشناسی ارشد چیست؟",
        "answer": "برای فارغ‌التحصیلی در دوره کارشناسی ارشد، دانشجویان باید تمامی واحدهای درسی خود را گذرانده و پایان‌نامه خود را با موفقیت دفاع کنند."
    },
    "c1_q5": {
        "question": "آیا امکان تغییر رشته تحصیلی وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند در صورت ارائه دلایل موجه و با تأیید گروه آموزشی، رشته تحصیلی خود را تغییر دهند."
    },
    "c1_q6": {
        "question": "آیا امکان تحصیل مجدد در همان مقطع تحصیلی وجود دارد؟",
        "answer": "بله، دانشجویانی که از یک مقطع فارغ‌التحصیل شده‌اند، می‌توانند با شرایط خاصی دوباره در همان مقطع ثبت‌نام کنند."
    },
    "c1_q7": {
        "question": "آیا امکان تحصیل همزمان در دو رشته وجود دارد؟",
        "answer": "خیر، دانشجویان نمی‌توانند به طور همزمان در دو رشته تحصیلی مشغول به تحصیل باشند."
    },
    "c1_q8": {
        "question": "چه زمانی باید برای تغییر رشته اقدام کرد؟",
        "answer": "دانشجویان باید در ابتدای تحصیل و در طول نیمسال اول برای تغییر رشته اقدام کنند."
    },
    "c1_q9": {
        "question": "آیا امکان شرکت در کلاس‌های جبرانی وجود دارد؟",
        "answer": "بله، در صورتی که دانشجو نمرات پایین کسب کرده یا به تقویت بیشتری نیاز داشته باشد، می‌تواند در کلاس‌های جبرانی شرکت کند."
    },
    "c1_q10": {
        "question": "آیا امکان درخواست تغییر گروه درسی وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند در ابتدای هر نیمسال و با موافقت گروه آموزشی، گروه درسی خود را تغییر دهند."
    },
    "c1_q11": {
        "question": "آیا امکان اخذ واحدهای جبرانی در دانشگاه دیگری وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند با کسب موافقت دانشگاه مبدا و مقصد، واحدهای جبرانی خود را در دانشگاه دیگری اخذ کنند."
    },

    # دسته‌بندی ۲. نمرات، اعتراض و تجدیدنظر
    "c2_q12": {
        "question": "چگونه می‌توان نمرات درسی را تجدیدنظر کرد؟",
        "answer": "دانشجویان می‌توانند تا حداکثر ۱۰ روز پس از اعلام نمرات، درخواست تجدیدنظر خود را به اداره آموزش ارائه دهند."
    },
    "c2_q13": {
        "question": "آیا امکان تجدیدنظر در نمرات پایان ترم وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند پس از اعلام نتایج، در صورت وجود اشتباه در تصحیح نمره، درخواست تجدیدنظر ارائه دهند."
    },
    "c2_q14": {
        "question": "آیا نمرات تجدیدنظر شده در معدل محاسبه می‌شود؟",
        "answer": "بله، پس از تجدیدنظر و اصلاح نمرات، نمرات نهایی در معدل کل دانشجو محاسبه خواهد شد."
    },
    "c2_q15": {
        "question": "چگونه می‌توان به نتایج امتحانات اعتراض کرد؟",
        "answer": "دانشجویان می‌توانند تا حداکثر ۱۰ روز پس از اعلام نتایج امتحانات، درخواست اعتراض به نتایج امتحانات خود را به اداره آموزش دانشگاه ارائه دهند."
    },
    "c2_q16": {
        "question": "آیا امکان تجدیدنظر در نمرات پایان‌نامه وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند پس از دفاع از پایان‌نامه، در صورت وجود اختلاف یا مشکل در نمره، درخواست تجدیدنظر کنند."
    },
    "c2_q17": {
        "question": "آیا نمرات تجدیدنظر شده به صورت رسمی در پرونده تحصیلی ثبت می‌شود؟",
        "answer": "بله، پس از تجدیدنظر در نمرات، نمرات جدید به صورت رسمی در پرونده تحصیلی دانشجو ثبت خواهد شد."
    },
    "c2_q18": {
        "question": "چگونه می‌توان به پایان‌نامه نمره بالاتری کسب کرد؟",
        "answer": "نمره پایان‌نامه بستگی به کیفیت کار علمی و دفاع موفق از آن دارد. دانشجویان باید به دقت کار تحقیقاتی خود را انجام دهند و از راهنمایی‌های استاد خود بهره‌مند شوند."
    },

    # دسته‌بندی ۳. مرخصی و انصراف از تحصیل
    "c3_q19": {
        "question": "آیا امکان اخذ مرخصی تحصیلی وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند با ارائه درخواست و موافقت دانشگاه، برای یک یا چند ترم مرخصی تحصیلی بگیرند."
    },
    "c3_q20": {
        "question": "آیا امکان تقاضای مرخصی تحصیلی برای مدت بیش از دو ترم وجود دارد؟",
        "answer": "در شرایط خاص، دانشجویان می‌توانند برای مدت بیش از دو ترم مرخصی تحصیلی بگیرند، اما این نیازمند تأیید شورای آموزشی است."
    },
    "c3_q21": {
        "question": "آیا مرخصی تحصیلی در دوره کارشناسی ارشد نیز محاسبه می‌شود؟",
        "answer": "بله، مرخصی‌های تحصیلی در دوره کارشناسی ارشد جزو مدت مجاز تحصیل محاسبه می‌شود."
    },
    "c3_q22": {
        "question": "چه زمانی می‌توان از دوره تحصیلی انصراف داد؟",
        "answer": "دانشجویانی که قصد انصراف از تحصیل دارند، باید درخواست رسمی خود را به دانشگاه ارائه دهند."
    },
    "c3_q23": {
        "question": "آیا امکان بازگشت به تحصیل پس از انصراف وجود دارد؟",
        "answer": "در موارد خاص و با تأیید دانشگاه، دانشجویانی که از تحصیل انصراف داده‌اند، می‌توانند مجدداً به تحصیل بازگردند."
    },
    "c3_q24": {
        "question": "آیا امکان تقاضای مرخصی تحصیلی در نیمسال اول وجود دارد؟",
        "answer": "خیر، دانشجویان نمی‌توانند در نیمسال اول تحصیل مرخصی بگیرند، مگر در شرایط خاص و با تأیید شورای آموزشی."
    },

    # دسته‌بندی ۴. واحدهای درسی و حذف و اضافه
    "c4_q25": {
        "question": "چگونه می‌توان واحدهای درسی را حذف یا اضافه کرد؟",
        "answer": "دانشجویان می‌توانند در ابتدای هر نیمسال تحصیلی و در بازه زمانی تعیین شده از سوی دانشگاه، واحدهای درسی خود را حذف یا اضافه کنند."
    },
    "c4_q26": {
        "question": "حداکثر تعداد واحدهایی که می‌توان در یک نیمسال اخذ کرد چقدر است؟",
        "answer": "حداکثر تعداد واحدهای درسی که می‌توان در هر نیمسال اخذ کرد، ۲۰ واحد است."
    },
    "c4_q27": {
        "question": "آیا امکان حذف اضطراری واحدهای درسی وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند در بازه زمانی مشخص و با تأیید استاد و گروه آموزشی، برخی از واحدهای درسی خود را حذف کنند."
    },
    "c4_q28": {
        "question": "چه زمانی می‌توان واحدهای درسی را اضافه کرد؟",
        "answer": "در ابتدای هر نیمسال تحصیلی، دانشجویان می‌توانند در بازه زمانی تعیین شده از سوی دانشگاه، واحدهای درسی خود را اضافه کنند."
    },
    "c4_q29": {
        "question": "آیا نمرات واحدهای درسی مهمان در معدل محاسبه می‌شود؟",
        "answer": "بله، نمرات واحدهای درسی گذرانده‌شده در دانشگاه مهمان در معدل کل دانشجو محاسبه خواهد شد."
    },
    "c4_q30": {
        "question": "آیا نمرات واحدهای جبرانی در معدل کل محاسبه می‌شود؟",
        "answer": "خیر، نمرات واحدهای جبرانی در معدل کل محاسبه نمی‌شوند."
    },
    "c4_q31": {
        "question": "آیا امکان حذف واحدهای درسی در طول نیمسال تحصیلی وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند با هماهنگی استاد و گروه آموزشی، برخی از واحدهای درسی خود را در طول نیمسال حذف کنند."
    },
    "c4_q32": {
        "question": "آیا نمرات جبرانی در معدل محاسبه می‌شود؟",
        "answer": "خیر، نمرات جبرانی در معدل کل محاسبه نمی‌شود."
    },
    "c4_q33": {
        "question": "آیا امکان اخذ واحدهای درسی از دانشگاه دیگری وجود دارد؟",
        "answer": "بله، در صورت موافقت دانشگاه مبدا و مقصد، دانشجویان می‌توانند برخی از واحدهای درسی خود را در دانشگاه‌های دیگر بگذرانند."
    },

    # دسته‌بندی ۵. امتحانات و پایان‌نامه
    "c5_q34": {
        "question": "آیا شرکت در امتحانات پایان ترم اجباری است؟",
        "answer": "بله، دانشجویان موظف به شرکت در تمامی امتحانات پایان ترم هستند. غیبت بدون عذر موجه به منزله نمره صفر برای آن درس خواهد بود."
    },
    "c5_q35": {
        "question": "چه شرایطی برای ارائه پایان‌نامه در دوره کارشناسی ارشد وجود دارد؟",
        "answer": "دانشجویان کارشناسی ارشد پس از اتمام واحدهای درسی باید پایان‌نامه خود را ارائه دهند و از آن با موفقیت دفاع کنند."
    },
    "c5_q36": {
        "question": "چه مدارکی برای دفاع از پایان‌نامه لازم است؟",
        "answer": "دانشجویان باید پس از تایید استاد راهنما و گروه آموزشی، مدارک مربوط به تکمیل پایان‌نامه خود را به اداره آموزش دانشگاه ارائه دهند."
    },
    "c5_q37": {
        "question": "آیا امکان درخواست تمدید مدت دفاع از پایان‌نامه وجود دارد؟",
        "answer": "بله، در صورت نیاز و با موافقت استاد راهنما و گروه آموزشی، مدت دفاع از پایان‌نامه می‌تواند تمدید شود."
    },
    "c5_q38": {
        "question": "آیا امکان تعویق امتحانات پایان ترم وجود دارد؟",
        "answer": "تعویق امتحانات تنها در شرایط خاص و با موافقت دانشگاه امکان‌پذیر است."
    },
    "c5_q39": {
        "question": "چه زمانی می‌توان برای دفاع از پایان‌نامه آماده شد؟",
        "answer": "دانشجویان پس از تکمیل تمامی واحدهای درسی و با تایید استاد راهنما می‌توانند برای دفاع از پایان‌نامه اقدام کنند."
    },
    "c5_q40": {
        "question": "آیا امکان دریافت گواهی موقت قبل از دفاع از پایان‌نامه وجود دارد؟",
        "answer": "خیر، مدرک موقت تنها پس از دفاع موفق از پایان‌نامه و تکمیل تمامی واحدهای درسی صادر می‌شود."
    },
    "c5_q41": {
        "question": "آیا نمرات دوره کارآموزی در معدل کل محاسبه می‌شود؟",
        "answer": "بله، نمرات دوره کارآموزی در معدل کل دانشجویان محاسبه می‌شود و جزئی از الزامات فارغ‌التحصیلی است."
    },
    "c5_q42": {
        "question": "چگونه می‌توان برای دوره کارآموزی ثبت‌نام کرد؟",
        "answer": "دانشجویان باید پس از گذراندن تعداد مشخصی از واحدهای درسی، برای دوره کارآموزی اقدام کنند و این دوره را با موفقیت به پایان برسانند."
    },
    "c5_q43": {
        "question": "آیا امکان اخذ واحدهای درسی جبرانی از دانشگاه دیگری وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند با کسب موافقت دانشگاه مبدا و مقصد، واحدهای جبرانی خود را در دانشگاه دیگری اخذ کنند."
    },
    "c5_q44": {
        "question": "آیا نمرات دروس تخصصی در ترم‌های ابتدایی قابل اخذ هستند؟",
        "answer": "خیر، دروس تخصصی معمولاً باید در ترم‌های پایانی دوره تحصیل گذرانده شوند."
    },
    "c5_q45": {
        "question": "آیا نمرات تجدیدنظر شده به صورت رسمی در پرونده تحصیلی ثبت می‌شود؟",
        "answer": "بله، پس از تجدیدنظر در نمرات، نمرات جدید به صورت رسمی در پرونده تحصیلی دانشجو ثبت خواهد شد."
    },

    # دسته‌بندی ۶. انتقالی و تحصیل در دانشگاه‌های دیگر
    "c6_q46": {
        "question": "آیا امکان انتقال از یک دانشگاه به دانشگاه دیگر وجود دارد؟",
        "answer": "بله، در شرایط خاص و با موافقت دانشگاه مبدا و مقصد، دانشجویان می‌توانند درخواست انتقال به دانشگاه دیگر دهند."
    },
    "c6_q47": {
        "question": "آیا امکان اخذ همزمان واحدهای درسی از دو دانشگاه وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند با موافقت دانشگاه مبدا و مقصد، برخی از واحدهای درسی خود را به صورت مهمان در دانشگاه دیگری اخذ کنند."
    },
    "c6_q48": {
        "question": "آیا نمرات دروس مهمان در معدل کل محاسبه می‌شود؟",
        "answer": "بله، نمرات واحدهای درسی گذرانده‌شده در دانشگاه مهمان در معدل کل دانشجو محاسبه خواهد شد."
    },

    # دسته‌بندی ۷. حذف و اضافه، واحدهای درسی و مشروط شدن
    "c7_q49": {
        "question": "چگونه می‌توان واحدهای درسی را حذف یا اضافه کرد؟",
        "answer": "دانشجویان می‌توانند در ابتدای هر نیمسال تحصیلی و در بازه زمانی تعیین شده از سوی دانشگاه، واحدهای درسی خود را حذف یا اضافه کنند."
    },
    "c7_q50": {
        "question": "حداکثر تعداد واحدهایی که می‌توان در یک نیمسال اخذ کرد چقدر است؟",
        "answer": "حداکثر تعداد واحدهای درسی که می‌توان در هر نیمسال اخذ کرد، ۲۰ واحد است."
    },
    "c7_q51": {
        "question": "چه زمانی می‌توان واحدهای درسی را اضافه کرد؟",
        "answer": "در ابتدای هر نیمسال تحصیلی، دانشجویان می‌توانند در بازه زمانی تعیین شده از سوی دانشگاه، واحدهای درسی خود را اضافه کنند."
    },
    "c7_q52": {
        "question": "آیا امکان حذف اضطراری واحدهای درسی پس از پایان مهلت حذف و اضافه وجود دارد؟",
        "answer": "خیر، پس از پایان مهلت حذف و اضافه، امکان حذف اضطراری واحدهای درسی وجود ندارد مگر در موارد خاص و با تأیید دانشگاه."
    },
    "c7_q53": {
        "question": "آیا نمرات جبرانی در معدل کل محاسبه می‌شود؟",
        "answer": "خیر، نمرات جبرانی در معدل کل محاسبه نمی‌شود. این دروس برای تقویت پایه علمی دانشجویان ارائه می‌شود."
    },
    "c7_q54": {
        "question": "آیا واحدهای درسی جبرانی در معدل محاسبه می‌شوند؟",
        "answer": "خیر، واحدهای جبرانی در معدل کل محاسبه نمی‌شوند. این دروس برای تقویت مهارت‌های تحصیلی دانشجویان تعیین می‌شوند."
    },
    "c7_q55": {
        "question": "شرایط مشروط شدن تحصیلی چیست؟",
        "answer": "دانشجویانی که معدل نهایی آن‌ها در یک نیمسال تحصیلی کمتر از ۱۲ باشد، مشروط می‌شوند."
    },

    # دسته‌بندی ۸. فارغ‌التحصیلی و گواهینامه‌ها
    "c8_q56": {
        "question": "شرایط دریافت گواهی موقت پایان تحصیلات چیست؟",
        "answer": "دانشجویانی که تمامی واحدهای درسی و الزامات آموزشی خود را به پایان رسانده‌اند، می‌توانند درخواست گواهی موقت پایان تحصیلات کنند."
    },
    "c8_q57": {
        "question": "آیا امکان دریافت گواهی موقت قبل از دفاع از پایان‌نامه وجود دارد؟",
        "answer": "خیر، گواهی موقت تنها پس از دفاع موفق از پایان‌نامه و تکمیل تمامی واحدهای درسی صادر می‌شود."
    },
    "c8_q58": {
        "question": "آیا امکان تغییر رشته در ترم‌های بالاتر وجود دارد؟",
        "answer": "خیر، تغییر رشته تنها در ترم‌های ابتدایی تحصیل و با تأیید گروه آموزشی امکان‌پذیر است."
    },
    "c8_q59": {
        "question": "چه زمانی دانشجویان مشمول محرومیت از تحصیل می‌شوند؟",
        "answer": "در صورتی که دانشجو نتواند نمرات و شرایط تحصیلی لازم را فراهم کند یا مقررات دانشگاه را رعایت نکند، ممکن است از ادامه تحصیل محروم شود."
    },
    "c8_q60": {
        "question": "آیا نمرات دروس تخصصی در معدل محاسبه می‌شود؟",
        "answer": "بله، نمرات دروس تخصصی در معدل کل دانشجویان محاسبه خواهد شد."
    },
    "c8_q61": {
        "question": "آیا امکان افزایش مدت تحصیل در دوره کارشناسی ارشد وجود دارد؟",
        "answer": "بله، در شرایط خاص و با موافقت شورای آموزشی، می‌توان مدت تحصیل در دوره کارشناسی ارشد را تا یک نیمسال دیگر تمدید کرد."
    },
    "c8_q62": {
        "question": "آیا امکان تمدید مدت تحصیل در دوره کارشناسی ارشد وجود دارد؟",
        "answer": "بله، در موارد خاص و با تأیید شورای آموزشی، مدت تحصیل در دوره کارشناسی ارشد ممکن است تا یک نیمسال دیگر تمدید شود."
    },
    "c8_q63": {
        "question": "آیا دانشجویان کارشناسی موظف به ارائه پایان‌نامه هستند؟",
        "answer": "خیر، پایان‌نامه مخصوص دوره‌های تحصیلات تکمیلی مانند کارشناسی ارشد است."
    },

    # دسته‌بندی ۹. کارآموزی و دوره‌های عملی
    "c9_q64": {
        "question": "آیا دانشجویان موظف به شرکت در دوره‌های آموزشی عملی هستند؟",
        "answer": "بله، برخی از رشته‌ها نیازمند گذراندن دوره‌های آموزشی عملی و کارآموزی هستند. دانشجویان باید این دوره‌ها را مطابق برنامه تحصیلی خود بگذرانند."
    },
    "c9_q65": {
        "question": "چه شرایطی برای گذراندن دوره کارآموزی وجود دارد؟",
        "answer": "دانشجویانی که دوره کارآموزی را می‌گذرانند، باید حداقل تعداد واحدهای درسی مربوطه را گذرانده و پس از آن برای گذراندن دوره کارآموزی اقدام کنند."
    },
    "c9_q66": {
        "question": "آیا نمرات کارآموزی در معدل کل محاسبه می‌شود؟",
        "answer": "بله، نمرات دوره کارآموزی در معدل کل دانشجویان محاسبه می‌شود و جزئی از الزامات فارغ‌التحصیلی است."
    },
    "c9_q67": {
        "question": "آیا امکان گذراندن دوره کارآموزی در دانشگاه دیگری وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند با موافقت دانشگاه مبدا و مقصد، دوره کارآموزی خود را در دانشگاه دیگری بگذرانند."
    },

    # دسته‌بندی ۱۰. آزمون‌ها و تجدید نظر
    "c10_q68": {
        "question": "آیا شرکت در امتحانات پایان ترم اجباری است؟",
        "answer": "بله، دانشجویان موظف به شرکت در تمامی امتحانات پایان ترم هستند. غیبت بدون عذر موجه به منزله نمره صفر برای آن درس خواهد بود."
    },
    "c10_q69": {
        "question": "چگونه می‌توان به نتایج امتحانات اعتراض کرد؟",
        "answer": "دانشجویان می‌توانند تا حداکثر ۱۰ روز پس از اعلام نتایج، درخواست اعتراض به نتایج امتحانات خود را به اداره آموزش دانشگاه ارائه دهند."
    },
    "c10_q70": {
        "question": "آیا امکان تجدیدنظر در نمرات پایان ترم وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند پس از اعلام نتایج، در صورت وجود اشتباه در تصحیح نمره، درخواست تجدیدنظر ارائه دهند."
    },
    "c10_q71": {
        "question": "آیا نمرات تجدیدنظر شده در معدل محاسبه می‌شود؟",
        "answer": "بله، پس از تجدیدنظر و اصلاح نمرات، نمرات نهایی در معدل کل دانشجو محاسبه خواهد شد."
    },
    "c10_q72": {
        "question": "آیا امکان تجدیدنظر در نمرات پایان‌نامه وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند پس از دفاع از پایان‌نامه، در صورت وجود اختلاف در نمره، درخواست تجدیدنظر کنند."
    },
    "c10_q73": {
        "question": "چه زمانی می‌توان به نتایج امتحانات پایان ترم اعتراض کرد؟",
        "answer": "دانشجویان می‌توانند تا حداکثر ۱۰ روز پس از اعلام نتایج امتحانات، درخواست اعتراض به نتایج خود را به دانشگاه ارائه دهند."
    },
    "c10_q74": {
        "question": "آیا امکان تمدید مدت تحصیل در دوره کارشناسی وجود دارد؟",
        "answer": "بله، در شرایط خاص و با موافقت دانشگاه، مدت تحصیل در دوره کارشناسی می‌تواند افزایش یابد."
    },
    "c10_q75": {
        "question": "آیا امکان تغییر نمرات پس از تجدید نظر وجود دارد؟",
        "answer": "بله، پس از تجدید نظر و تایید دانشگاه، نمرات جدید به صورت رسمی در پرونده تحصیلی دانشجو ثبت می‌شود."
    },
    "c10_q76": {
        "question": "آیا امکان حذف اضطراری واحدهای درسی پس از پایان مهلت حذف و اضافه وجود دارد؟",
        "answer": "خیر، پس از پایان مهلت حذف و اضافه، امکان حذف اضطراری واحدهای درسی وجود ندارد مگر در موارد خاص و با تأیید دانشگاه."
    },
    "c10_q77": {
        "question": "آیا نمرات دروس مهمان در معدل کل محاسبه می‌شود؟",
        "answer": "بله، نمرات دروس گذرانده‌شده در دانشگاه مهمان در معدل کل دانشجو محاسبه خواهد شد."
    },

    # دسته‌بندی ۱۱. ثبت‌نام و مقررات اداری
    "c11_q78": {
        "question": "چه زمانی باید برای ثبت‌نام در نیمسال جدید اقدام کرد؟",
        "answer": "دانشجویان باید در بازه زمانی مشخص‌شده توسط دانشگاه برای ثبت‌نام در نیمسال جدید اقدام کنند. عدم ثبت‌نام به موقع ممکن است منجر به محرومیت از تحصیل شود."
    },
    "c11_q79": {
        "question": "آیا می‌توان پس از اتمام مهلت ثبت‌نام برای نیمسال جدید اقدام کرد؟",
        "answer": "خیر، ثبت‌نام پس از اتمام مهلت تعیین‌شده توسط دانشگاه امکان‌پذیر نیست، مگر در شرایط خاص و با تأیید دانشگاه."
    },
    "c11_q80": {
        "question": "آیا امکان ثبت‌نام در دوره کارآموزی همزمان با تحصیل وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند با تأیید دانشگاه، همزمان با تحصیل در دوره‌های کارآموزی شرکت کنند."
    },
    "c11_q81": {
        "question": "آیا امکان تعویق امتحانات پایان ترم وجود دارد؟",
        "answer": "تعویق امتحانات تنها در شرایط خاص و با موافقت دانشگاه امکان‌پذیر است."
    },
    "c11_q82": {
        "question": "آیا نمرات تجدیدنظر شده به صورت رسمی در پرونده تحصیلی ثبت می‌شود؟",
        "answer": "بله، پس از تجدیدنظر در نمرات، نمرات جدید به صورت رسمی در پرونده تحصیلی دانشجو ثبت خواهد شد."
    },
    "c11_q83": {
        "question": "آیا امکان تقاضای مرخصی تحصیلی در نیمسال اول وجود دارد؟",
        "answer": "خیر، دانشجویان نمی‌توانند در نیمسال اول تحصیل مرخصی بگیرند، مگر در شرایط خاص و با تأیید شورای آموزشی."
    },
    "c11_q84": {
        "question": "آیا امکان ثبت‌نام در دو دانشگاه همزمان وجود دارد؟",
        "answer": "خیر، دانشجویان مجاز به ثبت‌نام همزمان در دو دانشگاه در یک دوره تحصیلی نیستند."
    },
    "c11_q85": {
        "question": "آیا امکان تمدید مدت تحصیل در دوره کارشناسی ارشد وجود دارد؟",
        "answer": "بله، در موارد خاص و با تأیید شورای آموزشی، مدت تحصیل در دوره کارشناسی ارشد ممکن است تا یک نیمسال دیگر تمدید شود."
    },
    "c11_q86": {
        "question": "آیا امکان شرکت در کلاس‌های آموزشی دانشگاه‌های دیگر وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند با موافقت دانشگاه مبدا و مقصد، به عنوان مهمان در کلاس‌های دانشگاه‌های دیگر شرکت کنند."
    },
    "c11_q87": {
        "question": "آیا امکان دریافت مدرک موقت قبل از دفاع از پایان‌نامه وجود دارد؟",
        "answer": "خیر، مدرک موقت تنها پس از دفاع موفق از پایان‌نامه و تکمیل تمامی واحدهای درسی صادر می‌شود."
    },

    # دسته‌بندی ۱۲. واحدهای جبرانی و دروس مهمان
    "c12_q88": {
        "question": "آیا واحدهای جبرانی در معدل محاسبه می‌شوند؟",
        "answer": "خیر، واحدهای جبرانی در معدل کل محاسبه نمی‌شوند. این دروس برای تقویت پایه علمی دانشجویان ارائه می‌شوند."
    },
    "c12_q89": {
        "question": "آیا نمرات دروس جبرانی در معدل کل محاسبه می‌شود؟",
        "answer": "خیر، نمرات دروس جبرانی در معدل کل محاسبه نمی‌شوند."
    },
    "c12_q90": {
        "question": "آیا امکان اخذ واحدهای جبرانی در دانشگاه‌های دیگر وجود دارد؟",
        "answer": "بله، دانشجویان می‌توانند با موافقت دانشگاه مبدا و مقصد، واحدهای جبرانی خود را در دانشگاه دیگری اخذ کنند."
    },
    "c12_q91": {
        "question": "آیا امکان اخذ واحدهای جبرانی پس از گذراندن تمام واحدهای اصلی وجود دارد؟",
        "answer": "بله، در صورت نیاز به تقویت پایه علمی، دانشجویان می‌توانند واحدهای جبرانی را پس از گذراندن دروس اصلی اخذ کنند."
    },
    "c12_q92": {
        "question": "آیا نمرات دروس مهمان در معدل محاسبه می‌شود؟",
        "answer": "بله، نمرات دروس گذرانده‌شده در دانشگاه مهمان در معدل کل دانشجو محاسبه خواهد شد."
    },
}

# تابع برای ایجاد منو با چند ردیف
def build_menu(buttons, n_cols, header_buttons=None, footer_buttons=None):
    menu = [buttons[i:i + n_cols] for i in range(0, len(buttons), n_cols)]
    if header_buttons:
        menu.insert(0, header_buttons)
    if footer_buttons:
        menu.append(footer_buttons)
    return menu

# هندلر برای دستور /start
def start(update: Update, context: CallbackContext):
    buttons = [InlineKeyboardButton(cat_name, callback_data=cat_id) for cat_id, cat_name in categories.items()]
    reply_markup = InlineKeyboardMarkup(build_menu(buttons, n_cols=1))  # تنظیم تعداد دکمه‌ها در هر ردیف
    update.message.reply_text('لطفاً یک دسته‌بندی را انتخاب کنید:', reply_markup=reply_markup)

# هندلر برای دریافت پاسخ به دسته‌بندی‌ها و سوالات
def button(update: Update, context: CallbackContext):
    query = update.callback_query
    query.answer()

    data = query.data

    if data in categories:
        # کاربر یک دسته‌بندی را انتخاب کرده است، نمایش سوالات مربوطه
        category_id = data
        category_name = categories[category_id]

        # جمع‌آوری سوالات مربوط به این دسته‌بندی
        buttons = []
        for key, pair in qa_pairs.items():
            if key.startswith(category_id):
                buttons.append(InlineKeyboardButton(pair["question"], callback_data=key))

        reply_markup = InlineKeyboardMarkup(build_menu(buttons, n_cols=1))
        query.edit_message_text(text=f'لطفاً یک سوال از دسته‌بندی "{category_name}" را انتخاب کنید:', reply_markup=reply_markup)
    elif data.startswith("c"):
        # کاربر یک سوال را انتخاب کرده است، نمایش پاسخ مربوطه
        answer = qa_pairs.get(data, {}).get("answer", "متاسفانه، نمی‌توانم به این سوال پاسخ دهم.")
        query.edit_message_text(text=answer)
    else:
        query.edit_message_text(text="متاسفانه، نمی‌توانم به این سوال پاسخ دهم.")

# هندلر برای سوالات غیر از دکمه‌ها (در صورت ارسال سوال دستی توسط کاربر)
def handle_message(update: Update, context: CallbackContext):
    question = update.message.text.strip()
    # جستجوی سوال در دیکشنری qa_pairs
    for pair in qa_pairs.values():
        if pair["question"].strip() == question:
            update.message.reply_text(pair["answer"])
            return
    update.message.reply_text("متاسفانه، نمی‌توانم به این سوال پاسخ دهم.")

# هندلر خطا
def error_handler(update: object, context: CallbackContext):
    print(f"خطا رخ داد: {context.error}")
    try:
        if isinstance(update, Update) and update.message:
            update.message.reply_text("متاسفانه، یک خطا رخ داده است. لطفاً دوباره تلاش کنید.")
    except:
        pass

# تابع اصلی برای اجرای بات
def main():
    updater = Updater(TELEGRAM_TOKEN, use_context=True)
    dp = updater.dispatcher

    # افزودن هندلرها
    dp.add_handler(CommandHandler("start", start))
    dp.add_handler(CallbackQueryHandler(button))
    dp.add_handler(MessageHandler(Filters.text & ~Filters.command, handle_message))

    # افزودن هندلر خطا
    dp.add_error_handler(error_handler)

    # شروع بات با استفاده از Polling
    updater.start_polling()
    print("بات در حال اجرا است...")
    updater.idle()

if __name__ == '__main__':
    main()
```
